## Problem 1: Smart Light Controller

- Problem Statement
- Create a class to control a smart light that can be turned ON or OFF and display its status.
- 🌍 Real-Time Use
- Smart home automation systems
- 📥 Sample Input
- Light Name: Bedroom Light
- Action: ON
- 📤 Expected Output
- Bedroom Light is ON

```python
class SmartLight:
    def __init__(self,name):
        self.name=name
        self.status="OFF"

    def control(self,action):
        if action=="ON":
            self.status="ON"
        elif action=="OFF":
            self.status="OFF"
        else:
            print("Invalid action")

    def display(self):
        print(self.name,"is",self.status)

name=input("Light Name:")
action=input("Action (ON?OFF):")

light=SmartLight(name)
light.control(action)
light.display()
```
