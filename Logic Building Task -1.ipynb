### __Q1__. User Login Check

- Problem Statement:
- Given a username and password, check whether login is successful.
- username = "admin"
- password = "1234"
- Requirements:
- Print "Login Successful" if both username and password match
- Otherwise print "Invalid Credentials"

```python
username=input("Enter your Username:")       ## user input for username
password=input("Enter your Password:")       ## user input for password

if username=="admin" and password=="1234":            ## and is used to compare the 2 values, when the two values are correct then it will accept
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### __Q2__. Pass / Fail Analyzer

- Problem Statement:
- Analyze student results.
- marks = [45, 78, 90, 33, 60]
- Requirements:
- A student passes if marks ≥ 50
- Count the total number of pass students
- Count the total number of fail students
- Print the final result clearly

```python
marks=[45, 78, 90, 33, 60]
passs=0
fail=0
for i in marks:
    if i>=50:                   ## student must have marks >=50
        # print("Student pass")
        passs+=1
    else:
        # print("Student fail")
        fail+=1
print("total number of pass students is:",passs)
print("total number of fail students is:",fail)
```

### __Q3__. Simple Data Cleaner

- Problem Statement:
- Clean and standardize user names.
- names = [" Alice ", "bob", " CHARLIE "]
- Requirements:
- Remove extra spaces
- Convert all names to lowercase
- Print the cleaned list

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]

for name in names:
    cleaned=name.strip()         ## used to remove the extra spaces
    cleaned=cleaned.lower()
    cleaned_names.append(cleaned)

print(cleaned_names)
```

### __Q4__. Message Length Analyzer

- Problem Statement:
- Analyze message sizes.
- messages = ["Hi", "Welcome to the platform", "OK"]
- Requirements:
- Print the length of each message
- Flag messages longer than 10 characters

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length=len(msg)
    print("Message:",msg)
    print("Length:",length)
    if length>10:
        print("Flag:Long message")
    else:
        print("Flag:Short message")
```

### __Q5__. Error Message Detector

- Problem Statement:
- Detect error messages from system logs.
- logs = ["INFO", "ERROR", "WARNING", "ERROR"]
- Requirements:
- Count the number of "ERROR" entries
- Print the total error count

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
    if i=="ERROR":
        count+=1
print("count of ERROR:",count)
```
